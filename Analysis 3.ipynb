## 2) What is the Average price of vehicle by fuel type and gearbox type.Give a plot.

```python
pd.crosstab(dd['fuelType'],dd['gearbox'], values = dd['price'], aggfunc = 'mean')
```

```python
pd.crosstab(dd['fuelType'],dd['gearbox'], values = dd['price'], aggfunc = 'mean').plot(kind = 'bar', xlabel = "Fuel Type", ylabel = 'Average Price')
```

## 3) What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot.

```python
pd.crosstab(dd['vehicleType'],dd['gearbox'], values = dd['powerPS'], aggfunc = 'mean')
```

```python
pd.crosstab(dd['vehicleType'],dd['gearbox'], values = dd['powerPS'], aggfunc = 'mean').plot(kind = 'bar', xlabel = "Type of Vehicle", ylabel = 'Average Power')
```

## 4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this

```python
pd.crosstab(dd['brand'], dd['vehicleType'] , values= dd['price'], aggfunc = 'mean')
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(pd.crosstab(dd['brand'], dd['vehicleType'] , values= dd['price'], aggfunc = 'mean'), annot = True, cmap='YlGnBu', fmt='.0f', cbar_kws={'label': "Average Price"})
plt.show()
```
